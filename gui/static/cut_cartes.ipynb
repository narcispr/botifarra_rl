```python
import imageio
import numpy as np
import matplotlib.pyplot as plt

# load image "baralla_completa.png" and display it
img = imageio.imread('baralla_completa.png')
plt.imshow(img)
plt.axis('off')
plt.show()
width = 208
height = 319
for rows in range(5):
    for cols in range(12):
        # cut the portion of the image corresponding to the card at (rows, cols)
        img_card = img[rows*height:(rows+1)*height, cols*width:(cols+1)*width]
        # save this piece of image as rows_cols.png
        imageio.imwrite(f'{rows*12 + cols}.png', img_card)
```
